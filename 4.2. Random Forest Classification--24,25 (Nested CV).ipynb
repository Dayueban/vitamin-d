```python
q1 = mf['OHV24D3'].quantile([0.10]).values
q3 = mf['OHV24D3'].quantile([0.90]).values
def V24D3_extreme(x):
    if (x <= q1):
        return(0)
    elif (x >= q3):
        return(1)
    else:
        return('NaN')
mf['V24D3_cat'] = mf['OHV24D3'].apply(V24D3_extreme)
```

## random forest classification 

### 24, 25

```python
# filter away samples with NaN values
Y = mf[mf.V24D3_cat != 'NaN']
Y.shape
```

```python
X = table.loc[Y.index]
X.shape
```

```python
# convert to numpy array
# X - rarefied OTU table; Y - vitamin D categories
y = pd.factorize(Y['V24D3_cat'])[0] # to get dtype 'int' instead of 'O'
X = X.as_matrix()
```

```python
m = np.logspace(2, 8, num=9, base=2, dtype=int)
p_grid = {"max_features": m,
          "min_samples_split": [2, 3, 7],
          "min_samples_leaf": [1, 3, 7],
          "bootstrap": [True, False]}

rf = RandomForestClassifier(n_estimators=1000)
```

```python
nested_CV_search = NestedCV(model=rf, params_grid=p_grid , outer_kfolds=5, inner_kfolds=5, n_jobs = -1,
                            cv_options={'metric':roc_auc_score, 
                                        'metric_score_indicator_lower':False, # higher auc score is better
                                        'randomized_search_iter':30, 
                                        'predict_proba':True})
nested_CV_search.fit(X=X,y=y)
```

```python
np.mean(nested_CV_search.outer_scores)
```
